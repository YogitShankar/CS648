```python
dictp = {}
for n in range(6,1000):
    count = 0
    for j in range(1,n):
        if (j**(n-1))%n != 1:
            count = count+1
    dictp[n]=count
```

```python
print(dictp)
```

```python
for n,value in dictp.items(): 
    if value == 0:
        print(n)
```

```python
import matplotlib.pyplot as plt

keys = list(dictp.keys())
values = list(dictp.values())

# Plotting
plt.scatter(keys, values, color='blue')
plt.xlabel('Keys')
plt.ylabel('Values')
plt.title('Key-Value Plot')
plt.grid(True)  # Adding grid for better visualization
plt.show()
```
